## les totaux

### les chiffres qui indiquent, pour chaque question, combien de fois une réponse paraît, sont regroupés dans un petit fichier .csv (disponible sur le même github que ce notebook). 

```python
# importer le .csv qui regroupe les chiffres

path_to_file = ("/Users/aidanairuser/Desktop/DATA/grand_debat/GD_sources_totaux.csv")
tots = pd.read_csv(path_to_file)
```

```python
tots
```

```python
# Jusqu'ici sont identifiés 1241 réponses 'collées' (nombre total de réponses = 153809) soit, moins qu'un pourcent.

tots.nb_réponses.sum()
print ((1241/153809)*100,"%")
```

```python
# regarder le décompte des réponses collées

grouped = tots.groupby("groupe").nb_réponses.sum()
grouped
```

```python
# visualiser

ax = grouped.plot(kind = 'bar', figsize = (10, 5), title = "Provenance des réponses 'types' à 5 questions (Grand débat: eco)")
ax.set_ylabel("Nombre de réponses", fontsize = 12)
ax.set_xlabel("Groupes (Asso, forum, groupe fb)", fontsize = 12)
ax
```
